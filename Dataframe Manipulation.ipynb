```python
import pandas as pd
```

```python
monthly = pd.read_csv('monthly_sales.csv')
```

```python
monthly['Year'] = monthly.apply(lambda x: x['Month, Year'][-2:], axis = 1)
monthly['Month'] = monthly.apply(lambda x: x['Month, Year'][0:3], axis = 1)
```

```python
monthly.head()
```

```python
burger_sums = monthly.sum()
burger_sums
```

```python
northeast = burger_sums['HM-NE'] + burger_sums['CF-NE'] + burger_sums['FF-NE']
southwest = burger_sums['HM-SW'] + burger_sums['CF-SW'] + burger_sums['FF-SW']
northwest = burger_sums['HM-SW'] + burger_sums['CF-SW'] + burger_sums['FF-SW']
southeast = burger_sums['HM-SW'] + burger_sums['CF-SW'] + burger_sums['FF-SW']
central = burger_sums['HM-SW'] + burger_sums['CF-SW'] + burger_sums['FF-SW']
```

```python
hamburgers = burger_sums['HM-NE'] + burger_sums['HM-SW'] + burger_sums['HM-NW'] + burger_sums['HM-SE'] + burger_sums['HM-C']
chicken = burger_sums['CF-NE'] + burger_sums['CF-SW'] + burger_sums['CF-NW'] + burger_sums['CF-SE'] + burger_sums['CF-C']
fish = burger_sums['FF-NE'] + burger_sums['FF-SW'] + burger_sums['FF-NW'] + burger_sums['FF-SE'] + burger_sums['FF-C']

all_burgers = hamburgers + chicken + fish
print(hamburgers/all_burgers)
print(chicken/all_burgers)
print(fish/all_burgers)


```

```python
all_sums = northeast + southeast + southwest + northwest + central
print(northeast / all_sums)
print(southwest / all_sums)
print(northwest / all_sums)
print(southeast / all_sums)
print(central / all_sums)
```
